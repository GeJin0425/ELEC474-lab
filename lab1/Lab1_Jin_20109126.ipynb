```python
imf = np.zeros((512,512,3),np.uint8)
for i in range (0,512):
    for j in range(0,512):
        (b,g,r) = img[i,j]
        b = 255-b
        g = 255-g
        r = 255-r
        imf[i,j] = (b,g,r)
plt.imshow(cv2.cvtColor(imf,cv2.COLOR_BGR2RGB))
plt.title("Lena Color negated")
```

```python
imf = np.zeros((512,512,3),np.uint8)
for i in range (0,512):
    for j in range(0,512):
        (b,g,r) = img[i,j]
        b = 255-b
        g = 255-g
        r = 255-r
        imf[i,j] = (b,g,r)
imf = cv2.cvtColor(imf,cv2.COLOR_BGR2GRAY)
plt.imshow(imf, cmap = 'gray')
plt.title("Lena gray negated")
```

```python
imf1 = np.zeros((512,512,3),np.uint8)
for i in range (0,512):
    for j in range(0,512):
        (b,g,r) = img[i,j]
        imf1[i,j] = (b,0,0)
        
plt.imshow(cv2.cvtColor(imf1,cv2.COLOR_BGR2RGB))
plt.title("Lena blue image")
```

```python
imf2 = np.zeros((512,512,3),np.uint8)
for i in range (0,512):
    for j in range(0,512):
        (b,g,r) = img[i,j]
        imf2[i,j] = (0,g,0)
        
plt.imshow(cv2.cvtColor(imf2,cv2.COLOR_BGR2RGB))
plt.title("Lena green image")
```

```python
imf3 = np.zeros((512,512,3),np.uint8)
for i in range (0,512):
    for j in range(0,512):
        (b,g,r) = img[i,j]
        imf3[i,j] = (0,0,r)
        
plt.imshow(cv2.cvtColor(imf3,cv2.COLOR_BGR2RGB))
plt.title("Lena red image")
```

```python
imf4 = np.zeros((512,512,3),np.uint8)
for i in range (0,512):
    for j in range(0,512):
        (b1,g1,r1) = imf1[i,j]
        (b2,g2,r2) = imf2[i,j]
        (b3,g3,r3) = imf3[i,j]
        b = b1+ b2+b3
        g = g1+g2+g3
        r = r1+r2+r3
        imf4[i,j] = (b,g,r)
plt.imshow(cv2.cvtColor(imf4,cv2.COLOR_BGR2RGB))
plt.title("Merged Lena image ")
```
